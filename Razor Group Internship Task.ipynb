```python
#AS WE CAN SEE THE PRODUCT COUNT IS THE SECOND LAST STRING ELEMENT IN THE SENTENCE SO WE CAN EXTRACT THAT AND UPDATE THE COLUMN.
```

```python
df1=df[~df['sellerproductcount'].isnull()]
df1

#REMOVING ALL NULL SELLER PRODUCT COUNT AS NO PRODUCT IS AVAILABLE 
```

```python
count_list=[]
def extract_count(col):
    for x in col:
        temp=[]
        temp=x.split()
        count=temp[-2]  #Second last element.
        count_list.append(count)
    return count_list
```

```python
extract_count(df1['sellerproductcount'])
```

```python
df1['Extracted Count']=count_list
```

```python
df2=df1[~df1['sellerratings'].isnull()]
df2
```

```python
percentage_rating=[]
count_of_positive_rating=[]
def rating_and_count(col):
    for x in col:
        temp=x.split('%')
        rating=int(temp[0])
        percentage_rating.append(rating)
        temp1=temp[1].split('(')
        temp2=temp1[1].split()
        count=int(temp2[0])
        count_of_positive_rating.append(count)
    return percentage_rating,count_of_positive_rating
        
        
```

```python
rating_and_count(df2['sellerratings'])
```

```python
percentage_rating
```

```python
count_of_positive_rating
```

```python
df2['Percentage rating']=percentage_rating
df2['COunt of Positive Rating']=count_of_positive_rating
```

```python
line=df2['sellerdetails'][0]
match=re.findall(r'[\w.+-]+@[\w-]+\.[\w.-]+', line)
match
```

```python
email_list=[]
def email(col):
    for x in col:
        match = re.findall(r'[a-z0-9\.\-+_]+@[a-z0-9\.\-+_]+\.[a-z]+', str(x))
        if len(match)==0:
            a=' '
        else:
            a=match[0]
            
        email_list.append(a)     
```

```python
email(df2['sellerdetails'])
```

```python
email_list
```

```python
df2['Emails'] = email_list
```

```python
df2.head(10)
```

```python
phone_numbers_list=[]
def phone_numbers(col):
    for x in col:
        match=re.findall(r'[\+\(]?[1-9][0-9 .\-\(\)]{8,}[0-9]', str(x))
        phone_numbers_list.append(match)
        
```

```python
phone_numbers(df2['sellerdetails'])
```

```python
phone_numbers_list
```

```python
df2['Phone Numbers']= phone_numbers_list
```

```python
#LOCATION IS THE LAST TWO STRING OF THE BUSINESS ADDRESS SO PARSING IT OUT USING SIMPLE SUBSTRING METHOD

location_list=[]
def location(col):
    for x in col:
        addr=str(x)
        loc=addr[len(addr)-2:len(addr)]
        location_list.append(loc)
```

```python
location(df2['businessaddress'])
```

```python
location_list
```

```python
df2['Location']=location_list
```
